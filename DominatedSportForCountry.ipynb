```python
import pandas as pd

df = pd.read_csv("athlete_events.csv")
```

```python
df = df[["Year", "Team", "Sex", "Event", "Medal","Sport"]].drop_duplicates()
#df = df[["Year", "Sport", "Country", "Gender", "Event", "Medal"]].drop_duplicates()
df = df.groupby(["Team", "Sport"])["Medal"].count().unstack()
```

```python

countries = [
    "United States", 
    "China", 
    "Russia", 
    "Germany","Canada",
    "Australia", "Japan", 
    "France", 
    "Italy" ,"Great Britain" ,"Soviet Union"   
]
country_colors = {
    "United States":"gray",
    "China":"red",
    "Russia":"yellow", 
    "Germany":"silver","Canada":"lightcoral",
    "Australia":"black", "Japan":"pink", 
    "France":"green",
    "Italy":"yellowgreen","Great Britain":"orange","Soviet Union":"brown"
}


Sport = [
    'Basketball', 'Football','Swimming','Badminton','Gymnastics','Table Tennis'
]
sm = df.loc[countries,'Swimming']
```

```python
sm = sm[::-1]
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("ticks")
sns.set_context("notebook", font_scale=1.2)
```

```python
colors = [country_colors[c] for c in sm.index]

plt.figure(figsize=(16,10))
sm.T.plot.pie( ax=plt.gca())

# Reverse the order of labels, so they match the data
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles[::-1], labels[::-1])

# Set labels and remove superfluous plot elements
plt.title("Medal won by country in Swimming")

sns.despine()
```
