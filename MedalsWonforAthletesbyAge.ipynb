```python
import pandas as pd
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("athlete_events.csv")
```

```python
df.groupby(['Age','Medal']).size().unstack().plot(kind='bar',stacked=True,figsize=(16,10))
plt.title('Medals Won for Athletes by Age')
plt.show()
```
