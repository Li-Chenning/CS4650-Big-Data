```python
import pandas as pd
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("athlete_events.csv")
```

```python
Medals = df[(df.Medal == 'Gold')|(df.Medal == 'Silver')|(df.Medal == 'Bronze')]
nd = Medals['Sport'][Medals['Age'] > 50]
plt.figure(figsize=(20, 10))
sns.countplot(nd)
plt.title('Medals Won for Athletes Over 50')
```

```python
Medals = df[(df.Medal == 'Gold')|(df.Medal == 'Silver')|(df.Medal == 'Bronze')]
nd = Medals['Sport'][(Medals['Age'] > 10) &(Medals['Age'] < 15) ]
plt.figure(figsize=(20, 10))
sns.countplot(nd)
plt.title('Medals Won for Athletes between 10 and 15')
```
